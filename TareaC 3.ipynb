```python
from pylab import *
import scipy, math
%matplotlib inline
```

```python
def binomial(x,p,q,N):
    y = ((scipy.math.factorial(N))/((scipy.math.factorial(x))*(scipy.math.factorial(N-x))))*(p**x)*(q**(N-x))
    return y
```

```python
for i in arange(5):
    x = i 
    plot(x,binomial(x,1./3,2./3,5), 'bo')
    
for i in arange(10):
    x = i 
    plot(x,binomial(x,1./3,2./3,10), 'ro')
    
for i in arange(100):
    x = i 
    plot(x,binomial(x,1./3,2./3,100), 'ro')
    
xlim = 50
ylim = 0.4
```

```python
def graf(N):
    x = arange(N)
    y = zeros(N)
    for i in range(N):
        y[i] = binomial(i,1./3,2./3,N)

    plot(x,y)
```

```python
graf(5)
graf(10)
graf(100)
```

```python
def normal(x,p,q,N):
    sigma = N*p*q
    mu = N*p
    y = (1./(sqrt(2*pi*(sigma**2))))*exp(-((x-mu)**2/(2*sigma**2)))
    return y
```

```python
def graf2(N):
    x = arange(N)
    y = zeros(N)
    for i in range(N):
        y[i] = normal(i,1./3,2./3,N)

    plot(x,y)
```

```python
graf2(5)
graf2(10)
graf2(100)
```
